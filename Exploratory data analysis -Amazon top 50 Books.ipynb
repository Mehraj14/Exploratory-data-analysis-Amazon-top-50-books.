# Amazon Top 50 best selling books 2009-2019 (Exploratory data analysis)

 The Note book is about the amazon top 50 bestselling books from 2009-2019. It contains 550 books, data has been categorized into fiction and non-fiction using good reads.
 The data set include several variables such as Name of the book, Author of the book,User Rating, Number of reviews, Price of the books, Year and the category of the books. 
In the exploratory data analysis we will be analysing each variable and try to find the different insights from the data.


```python
#Importing all the libraries 
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#Reading the data
df=pd.read_csv("bestsellers with categories (4).csv")
df.head()
```

```python
#Finding the types of the data in the dataset
df.dtypes
```

```python
# Getting the numbers of rows and columns
df.shape
```

```python
#Summary of the data frame
df.info()
```

```python
#Checking the missing value 
df.isnull().sum()
```

```python
#Visulaizing the missing value
sns.heatmap(df.isnull())
```

There is no missing value in the data set.

```python
#To finding any duplicate value in the data set
df.duplicated().sum()
```

There is no duplicate value in the data set

```python
#Desciribing the overall data set
df.describe()
```

# Univariate , Bivariate and Multivariate analysis of the variables.

# Genre

```python
#Countning the types of genre in data set
df.Genre.value_counts()
```

As we can see that Non fiction books are higher than fiction books.


```python

#For visulaizing the Gnere we need to convert the categorical varilable into numerical form
df.Genre=pd.Categorical(df.Genre,['Non Fiction','Fiction'])
```

```python
df.Genre=df.Genre.cat.codes
```

```python
df.head()
```

```python
#Visulaizing Genre into barplot
sns.distplot(df.Genre,color='blue')
```

From the distplot it's clearly visible that non-fiction books are much higher than fiction books.


# User rating

```python
# To understand the distribution of the user rating
sns.kdeplot(df['User Rating'],color='red')
plt.xlabel('Rating distribution')
plt.ylabel('Frequency')
plt.show()
```

 As we can see from the plot that most of the user ratings are from 4.00 to 5.00 and maximum number of user rating are in 4.75.

# Price 

```python
# Understanding the price distribtion of the data
sns.distplot(df.Price,color='green')
plt.xlabel('Price')
plt.ylabel('frequency')
plt.show()
```

Most of the books price are in the range of 0-20. There are few books which prices were higher.

# Reviews

```python
# To undersatnd the distribution of the review of the books.
sns.distplot(df['Reviews'],color='green')
plt.xlabel('Reviews')
plt.ylabel('frequency')
plt.show()
```

Most of the books got review from 0 to 10000 few of the books got reviews from 60000 TO 80000.

# Bivariarte analysis


# The year with highest user rating

```python
# For visulizig the year with highest user rating
sns.lineplot(x='Year',y='User Rating',data=df)
```

  As we can seethat In the year of 2019 the user rating was highest.
  In the year of 2012 the user rating was lowest.

# The year with highest review

```python
#The year with highest review
sns.barplot(x='Year',y='Reviews',data=df)
```

 In the year of 2014 highest numbers of reviews came in the books and in the year of 2009 the lowest number of review were recorded.

# Price variation with time

```python
#Price variation through time
sns.lineplot(x='Year',y='Price',data=df,color='red')
```

The prices of the books were declined with time. When the books were launched first there prices were higher and with time their prices decline.

# Highest review book

```python
#Finding out the highest review book
highest_review=df.nlargest(20,['Reviews'])
sns.barplot(y='Name',x='Reviews',data=highest_review,color='red')
```

The highest review book was Where the crawdads sing it has more than 80000 reviews.


# Lowest review book

```python
# Finding the lowest review book
lowest_review=df.nsmallest(20,['Reviews'])
sns.barplot(x='Reviews',y='Name',data=lowest_review,color='violet')
```

The lowest review book was Dvine soul mind and body healig and transmission system it has less than 100 review were there in that book.

# Genre and Reviews

```python
#Genre and reviews
#To understand the realtionship between genre and reviews
sns.barplot(x='Genre',y='Reviews',data=df,color='green')
```

The most number of reviews were in the fiction books.

# Genre and price

```python
#Visulaizing the Genre and price
sns.barplot(x='Genre',y='Price',data=df,color='green')
```

The price of the fiction books were much higher than the non-fiction books.

# Most expensive book


```python
#First sort the books by it's price than ploting into bar diagram
most_expensive=df.sort_values(by=['Price'],ascending=False).head(10)
sns.barplot(x='Price',y='Name',data=most_expensive,color='red')
```

The most expensive book was Diagnostic and statistical manual of mental disorders. Price was 120 and where as the average price of the books were in the rage of 10-20 dollar.

# Free Books
The books which price were 0 amazon distributing it free of cost.

```python
#Sorting values by price and ploting intombar diagram
free_books=df[df['Price']==0].head(10)
free_books.sort_values(by=['User Rating'],ascending=False)
sns.barplot(y='Name',x='User Rating',data=free_books,color='red')
```
